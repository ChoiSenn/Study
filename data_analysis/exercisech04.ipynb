## 3. plot() 함수의 선 그리기 기능

<br>



```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-20, 20)  # -20에서 20 사이의 수를 1의 간격으로 생성
y1 = 2 * x 
y2 = (1/3) * x ** 2 + 5 
y3 = -x ** 2 - 5

plt.plot(x, y1, 'g--', x, y2, 'r^-', x, y3, 'b*:')  # 빨간색 점선, 녹색 실선과 세모기호, 파랑색 별표와 점선으로 함수 표현
plt.axis([-30, 30, -30, 30])  # 그림 그릴 영역 지정
```

## 6. Figure, axes

<br>

subplot() 함수는 Figure 객체와 Axes 객체를 반환한다.

<br>

Figure 객체는 축과 그래픽, 텍스트 등의 그리기 객체들을 포함하는 컨테이너 객체이다.<br>
Axes 객체는 그림을 그릴 수 있는 테두리 상자 모양의 공간이다.<br>
(Figure 객체 안에 Axes 객체가 있고, Axes 객체 안에 그래프가 그려짐.)

```python
fig, ax = plt.subplots(2, 2)  # 2행 2열의 그림 그리는 공간 생성

x = np.random.randn(100)  # 정규 분포를 가지는 데이터
y = np.random.randn(100)  # 정규 분포를 가지는 데이터  
ax[0, 0].scatter(x, y)  # 산점도 그림

x = np.arange(10)  # 0~9 생성
y = np.random.uniform(0, 10, 10)  # 균일분포값 생성
ax[0, 1].bar(x, y)  # 막대 차트 그림

x = np.linspace(0, 10, 100)
y = np.cos(x)
ax[1, 0].plot(x, y)  # 실선으로 함수 그림

z = np.random.uniform(0, 1, (5, 5))
ax[1, 1].imshow(z)  # 분포를 2D 이미지로 그림
plt.style.use('default')
```

## 7. subplot()의 고급 기능

<br>

```python
fig, ax = plt.subplots(2, 3)

for i in range(2):  # 2행
  for j in range(3):  # 3행
    ax[i, j].text(0.3, 0.5, str((i, j)), fontsize = 11)  # 텍스트만 표시
```

```python
fig, ax = plt.subplots(2, 3)

grid = plt.GridSpec(2, 3, wspace = 0.4, hspace=0.3)  # wspace, hspace는 각각 수평, 수직 여백 표시. 0으로주면 격자 그림 사이의 간격 사라짐

x = np.random.randn(100)
y = np.random.randn(100)
plt.subplot(grid[0, 0]).scatter(x, y)

x = np.arange(10)  # 0~9 생성
y = np.random.uniform(0, 10, 10)  # 균일분포값 생성
plt.subplot(grid[0, 1:]).bar(x, y)  # 막대 차트 그림

x = np.linspace(0, 10, 100)
y = np.cos(x)
plt.subplot(grid[1, :2]).plot(x, y)  # 실선으로 함수 그림

z = np.random.uniform(0, 1, (5, 5))
plt.subplot(grid[1, 2]).imshow(z)  # 분포를 2D 이미지로 그림
```

```python
fig, ax = plt.subplots(3, 3)

grid = plt.GridSpec(3, 3, wspace = 0.4, hspace=0.3)  # wspace, hspace는 각각 수평, 수직 여백 표시. 0으로주면 격자 그림 사이의 간격 사라짐

x = np.random.randn(100)
y = np.random.randn(100)
plt.subplot(grid[0, 0:]).scatter(x, y)

x = np.arange(10)
y = np.arange(10)
plt.subplot(grid[1, :2]).plot(x, y, 'r--') 

z = np.random.uniform(0, 1, (5, 5))
plt.subplot(grid[1:, 2]).imshow(z) 
x = np.linspace(0, 10, 100)
y = np.cos(x)
plt.subplot(grid[2, :1]).plot(x, y)

x = np.arange(10)
y = np.random.uniform(1, 10, 10)
plt.subplot(grid[2, 1]).bar(x, y)
```

<br>

```python
import matplotlib.pyplot as plt
import numpy as np

data = [5, 4, 6, 11]
clist = ['cyan', 'gray', 'orange', 'red']
explode = [.06, .07, .08, .09]  # 중심에서 벗어나는 정도를 백분율로 지정
plt.pie(data, autopct = '%.2f', colors=clist, labels = clist, explode = explode)  # autopct는 부채꼴 안에 표시될 숫자의 형식 지정(소수점 아래 둘째 자리까지). labels는 영역별 레이블값 지정
```

<br><br>

데이터 값이 클 경우와 작은 경우에 대하여 서로 다른 색상을 사용하여 데이터의 분포를 쉽게 이해할 수 있다.

## 11. 히스토그램을 이용한 정규 분포 함수와 확률 밀도 함수 그리기

<br>

정규 분포란, 연속 확률 분포의 하나이며 평균이 0이고 분산이 1인 정규 분포를 표준 정규 분포라고 한다.

<br>


```python
f1 = np.random.normal(loc = 0, scale = 1, size = 10000)  # 10000개의 난수를 생성하고 이를 이용해 히스토그램 그리기(평균이 0이고 분산이 1인 표준 정규 분포)
f2 = np.random.normal(loc = 3, scale = .5, size = 10000)  # (평균이 3.0이고 분산이 0.5)

plt.hist(f1, bins = 200, color = 'red', alpha = .7, label = 'loc = 0, scale = 1')
plt.hist(f2, bins = 200, color = 'blue', alpha = .5, label = 'loc = 3, scale = .5')
```

<br>

hist()함수의 density 인자가 True로 설정되어 있으면 히스토그램을 그리며 동시에 확률 밀도 값을 반환한다.<br>
따라서 이 데이터를 바탕으로 norm.fit() 함수를 사용할 경우, 평균(mu)과 표준편차(std)를 얻을 수 있다. 이 평균과 표준편차를 바탕으로 확률 밀도 함수를 그릴 수 있다.

## 12. 상자 수염 그리기

<br>

상자 수염 그림은 수치 데이터를 표현하는 그래프의 한 종류이다. 흔히 상자차트(박스 프롯)라고도 한다.<br>
이 그래프는 최대, 최소, 중간값과 사분위 수 등의 통계량을 하나의 상자 그림으로 가시화하는 방법이다.

<br>


```python
rand_data = np.random.randn(100)  # 넘파이의 난수 모듈에 있는 정규 분포에 따르는 난수 100개를 생성

plt.boxplot(rand_data)
```

<br>

- 중앙값: 상자 중앙에 나타난 주황색 선. 데이터의 정 가운데 순위 값을 말함.<br>
- Q3: 중앙값을 기준으로 전체의 상위 25%의 수준.<br>
- Q1: 중앙값을 기준으로 전체의 하의 25%의 수준.<br>
- 사분 범위(IQR): Q3-Q1
- 위스커(수염): 상자 아래 위로 그어진 직선을 위/아래에서 막는 수평선. 이상치 자료 값의 경계 표시.<br>
- 상단 수염: Q2 + 1.5 * IQR 값보다 작은 데이터 중 가장 큰 값<br>
- 하단 수염: Q2 - 1.5 * IQR 값보다 큰 데이터 중 가장 작은 값<br>
- 이상치: 위스커를 넘어선 곳에 있는 점.

<br>

즉, 아랫쪽 위스커에서 윗쪽 위스커까지의 범위가 대부분의 데이터가 분포하는 범위이다.

<br>


```python
np.random.seed(85)

data1 = np.random.normal(100, 10, 200)  # 평균이 100이고 분산이 10
data2 = np.random.normal(100, 40, 200)  # 평균이 100이고 분산이 40
data3 = np.random.normal(80, 40, 200)  # 평균이 80이고 분산이 40
data4 = np.random.normal(80, 60, 200)  # 평균이 80이고 분산이 60

plt.boxplot([data1, data2, data3, data4])  # 평균이 작을수록 중앙값이 낮아지며, 분산이 클수록 상자의 길이가 늘어난다.
```
