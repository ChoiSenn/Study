## 4-1. 다음의 기능을 넘파이를 사용하여 구현하시오.

### 1) 대표적인 주기함수인 사인 함수를 이용하여 다음 그림과 같은 히트맵을 그려라. 격자의 크기는 10 X 10이며, 사인 함수의 입력값의 구간은 0에서 파이 사이이다.

<br>


```python
import matplotlib.pyplot as plt
import numpy as np
import math

data = np.linspace(start = 0, stop = np.pi, num = 100)  # 0부터 pi까지, 100개의 배열 만들기
data = data.reshape(10, 10)  # 배열을 10 X 10 크기의 2차원 배열로
plt.imshow(np.sin(data))  # 히트맵 그리기
plt.colorbar()
```

*linspace()를 이용하여 0부터 np.pi 즉 파이값 사이의 순차적인 배열을 100개 만들어 data에 저장한다.<br>
data를 10 X 10 크기의 2차원 배열로 만들기 위해 reshape를 이용한다.<br>
해당 data를 sin()을 이용해 사인 함수로 히트맵을 그린다.*

<br>

*사인 함수를 이용해 만들었기 때문에 가운데 부분은 노란색(큰 값), 가장자리 부분은 진한 색(낮은 값)으로 보인다.*

<br>

### 2) 사인 함수를 사용하여 100 X 100 크기의 격자 공간에 패턴이 나타나도록 코드를 수정하라.

<br>

```python
import matplotlib.pyplot as plt
import numpy as np
import math

data = np.linspace(start = 0, stop = np.pi, num = 10000)  # 0부터 pi까지, 10000개의 배열 만들기
data = data.reshape(100, 100)  # 배열을 100 X 100 크기의 2차원 배열로
plt.imshow(np.sin(data))  # 히트맵 그리기
plt.colorbar()
```

*위 문제와 같지만, linspace로 배열 생성 시 크기를 10000으로 지정하였고, reshape 시에도 100 X 100 크기의 2차원 배열로 가공해주었다.*

<br>

*이 역시 사인 함수를 이용하여 만들었기 때문에 가운데 부분만 높은 값을 띄는 것을 확인할 수 있다.*

## 4-2. 난수를 1000개 발생시켜 임의의 2차원 좌표를 생성하여 그려보자. 이때 좌표의 x와 y 값은 표준 정규 분포를 따르게 한다. 그러면 생성되는 좌표는 원점 (0. 0)에 밀집한 모양을 가질 것이다.

<br>

### 1) scatter() 함수 내에 불투명도를 의미하는 alpha 키워드 매개변수를 0.01로 두고 그림을 그려보자.

<br>

```python
n = 1000  # 난수 개수 설정
x = np.random.randn(n)  # x 좌표 값 n개 생성
y = np.random.randn(n)  # y 좌표 값 n개 생성

plt.scatter(x, y, alpha = 0.01)
```

*n, 즉 난수의 개수를 1000개로 설정하고, n개 만큼 x축과 y축 좌표를 생성해준다.<br>
scatter()를 이용하여 x와 y 값으로 산점도 그래프를 그리고 alpha 인자를 이용해 투명도를 지정한다.*

<br>

*표준 정규 분표를 좌표로 이용하여 산점도 그래프를 작성하였기 때문에, 전체적으로 분포가 0, 0좌표에 몰려있는 것을 확인 할 수 있다.*

<br>

```python
n = 1000  # 난수 개수 설정
x = np.random.randn(n)  # x 좌표 값 n개 생성
y = np.random.randn(n)  # y 좌표 값 n개 생성

plt.scatter(x, y, alpha = 0.1, color = 'magenta')
```

*코드는 위와 같으며, alpha를 0.1로 주어 약간 진하게 표현하였고 color를 마젠타로 주어 색상을 지정해주었다.*

<br>

*위와 동일하게 산점도가 0, 0좌표에 집중되어 있는 것을 확인 할 수 있다. alpha로 투명도를 주었기 때문에 값이 많이 위치한 부분은 더 진하게 표시되어 한 눈에 알아보기 쉽다.*

<br>

## 4-3. 정규 분포를 가지는 10000개의 난수를 생성하여 다음과 같은 f1, f2 다차원 배열을 만들고 히스토그램을 생성한다.

<br>



```python
f1 = np.random.normal(loc = 0, scale = 1, size = 10000)
f2 = np.random.normal(loc = 3, scale = .5, size = 10000)

fig, ax = plt.subplots(2, 2)
grid = plt.GridSpec(2, 2, wspace = 0.2, hspace = 0.2)

plt.subplot(grid[0, 0]).hist(f1, bins = 100, color = 'red', alpha = 0.5)
plt.title('Red histogram')
plt.subplot(grid[0, 1]).hist(f2, bins = 100, color = 'blue', alpha = 0.5)
plt.title('Blue histogram')
plt.subplot(grid[1, 0:]).hist(f1, bins = 100, color = 'red', alpha = 0.5, label = "loc = 0, scale = 1")
plt.legend()
plt.subplot(grid[1, 0:]).hist(f2, bins = 100, color = 'blue', alpha = 0.5, label = "loc = 3, scale = .5")
plt.legend()
```

*우선 제시된대로 정규 분포를 가지는 10000개의 난수를 생성하여 배열 f1, f2를 생성한다.<br>
subplots()를 이용하여 2 X 2 크기의 그래프 판을 만들고 wspace와 hspace로 상하좌우 여백 크기를 지정한다.<br>
각각 [0, 0], [0, 1], [1, 0:]좌표에 그래프를 그린다.<br>
[0, 0], [0, 1]의 그래프에는 title을 이용하여 제목을 붙여준다.<br>
[1, 0:]에는 그래프 두 개를 동시에 그리고, label을 이용하여 레이블을 붙여준다.*

<br>

*왼쪽 상단의 그래프는 loc이 0, 즉 좌표 0이 평균인 정규 분포 그래프이며 scale은 1 정도의 분산을 의미한다.<br>
오른쪽 상단의 그래프는 loc이 3, 즉 좌표 3이 평균인 정규 분포 그래프이며 scale은 분산이 0.5라는 것을 알 수 있다.<br>
하단의 그래프를 통해 차이를 더욱 확실히 확인할 수 있다.<br>
푸른 색 히스토그램은 평균이 3이므로 붉은 그래프보다 훨씬 오른쪽에 위치해있다. 또한, 분산이 1인 붉은 색 그래프가 분산이 0.5인 푸른 색 그래프보다 값들이 넓게 퍼져있다는 사실을 확인할 수 있다.*
